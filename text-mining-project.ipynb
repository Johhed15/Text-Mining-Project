```python
from typing import List
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("juierror/flan-t5-text2sql-with-schema-v2")
model = AutoModelForSeq2SeqLM.from_pretrained("juierror/flan-t5-text2sql-with-schema-v2")

```

```python

def get_prompt(tables, question):
    prompt = f"""convert question and table into SQL query. tables: {tables}. question: {question}"""
    return prompt

def prepare_input(question: str, tables: dict[str, List[str]]):
    tables = [f"""{table_name}({",".join(tables[table_name])})""" for table_name in tables]
    tables = ", ".join(tables)
    prompt = get_prompt(tables, question)
    input_ids = tokenizer(prompt, max_length=1000, return_tensors="pt",truncation=True).input_ids
    return input_ids

def inference(question: str, tables: dict[str, List[str]]) -> str:
    input_data = prepare_input(question=question, tables=tables)
    input_data = input_data.to(model.device)
    outputs = model.generate(inputs=input_data,  max_length=1000)
    result = tokenizer.decode(token_ids=outputs[0], skip_special_tokens=True)
    return result
```

```python
from fuzzywuzzy import fuzz
from fuzzywuzzy import process


def get_closest_country(input_country, country_list):
    # Find the closest match to the input country from the country list
    best_match = process.extractOne(input_country, country_list)
    return best_match[0]  # Return the closest country name

def extract_data(sql):
    # From the genereated SQL retrieve the data and return both data and SQL query
    conn = sqlite3.connect("database_life_exp.db")
    cursor = conn.cursor() # open database

    # try the sql call
    try:
        if 'Country =' in sql: # Extract country from the SQL query (if applicable)
            cursor.execute("SELECT DISTINCT Country FROM Life_Expectancy")
            countries = [row[0] for row in cursor.fetchall()] # all countries in the data

            start = 0
            while 'Country = "' in sql[start:]:
                # Find the start and end of the current country in the SQL string
                country_start = sql.find('Country = "', start) + len('Country = "')
                country_end = sql.find('"', country_start)
                
                # Extract the misspelled country
                misspelled_country = sql[country_start:country_end]
                
                # Correct the country name
                correct_country = get_closest_country(misspelled_country, countries)
                
                # Replace the misspelled country with the corrected one
                sql = sql[:country_start] + correct_country + sql[country_end:]
                
                # Update the start position for the next occurrence
                start = country_end
            
        cursor.execute(sql) 
        rows = cursor.fetchall()
        
        # Check if rows are empty or contain only (None,)
        if rows == [(None,)]:
            rows = 'No data found for the query.'
        
    except Exception as e:
        print(f"Error: {e}") 
        rows  = 'The SQL query didnt go through'
    finally:
        conn.close()

    return sql,rows # return the sql and fetched data

    
```

```python
from datasets import load_dataset, concatenate_datasets
spider = load_dataset("Spider")
```

```python
tokenizer_spider = AutoTokenizer.from_pretrained("google/flan-t5-base")
# base model
```

```python
# declare model
model_spider = AutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-base")
model_spider  = model_spider.cuda()
```

```python
# prefix to work on my dataset

table_name = 'Life_expectancy'
quest = 'Question'

def preprocess_function(db: dict[str, any]):
    """
    Preprocess Spider dataset for T5 fine-tuning.
    
    Args:
        db (dict): A batch of examples from the Spider dataset.
        
    Returns:
        dict: Processed inputs and targets suitable for the T5 model.
    """
    inputs = []
    targets = []
    
    for i in range(len(db["db_id"])):
        # Extract relevant fields
        question = db["question"][i].strip()
        db_id = db["db_id"][i]
        query = db["query"][i].strip()
        
        # Combine question and database identifier into a single input
        input_str = f"convert question and table into SQL query. Question: {question} | Table: {db_id}"
        
        # Add the formatted input and target to respective lists
        inputs.append(input_str)
        targets.append(query)
    
    # Tokenize the input and target sequences
    model_inputs = tokenizer_spider(inputs, text_target=targets, max_length=1000, truncation=True)
    return model_inputs
```

```python
# preprocess
train_dataset = spider["train"].map(preprocess_function, batched=True,remove_columns=['query_toks', 'query_toks_no_value', 'question_toks','db_id', 'query', 'question'])
val_dataset = spider["validation"].map(preprocess_function, batched=True,remove_columns=['query_toks', 'query_toks_no_value', 'question_toks','db_id', 'query', 'question'])
```

```python
# collator to for padding during batching
from transformers import DataCollatorForSeq2Seq
from transformers import Seq2SeqTrainingArguments, Seq2SeqTrainer
data_collator = DataCollatorForSeq2Seq(tokenizer=tokenizer_spider, model=model_spider)
```

```python
from sklearn.metrics import f1_score

import os
os.environ["TOKENIZERS_PARALLELISM"] = "true"  # Enable tokenizer parallelism
# Disable W&B logging
os.environ["WANDB_DISABLED"] = "true"
# In your training arguments
training_args = Seq2SeqTrainingArguments(
    output_dir="./results",
    eval_strategy="steps",
    eval_steps=381,
    save_strategy="steps",
    save_steps=381,
    learning_rate=1e-4,
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    gradient_accumulation_steps=8,
    warmup_ratio=0.01,
    weight_decay=0.01,
    save_total_limit=2,
    num_train_epochs=10,
    fp16=True,
    predict_with_generate=True,
    lr_scheduler_type="cosine",
    dataloader_num_workers=4,  # Increase the number of workers for parallelism
    greater_is_better=False,
    metric_for_best_model="eval_loss" # changing metric
)
```

```python
#  trainer
trainer = Seq2SeqTrainer(
    model=model_spider,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    processing_class=tokenizer_spider,
    data_collator=data_collator
) 
```

```python
trainer.train()
```
