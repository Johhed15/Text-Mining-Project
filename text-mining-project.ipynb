```python
from typing import List
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("juierror/flan-t5-text2sql-with-schema-v2")
model = AutoModelForSeq2SeqLM.from_pretrained("juierror/flan-t5-text2sql-with-schema-v2")

```

```python

def get_prompt(tables, question):
    prompt = f"""convert question and table into SQL query. tables: {tables}. question: {question}"""
    return prompt

def prepare_input(question: str, tables: dict[str, List[str]]):
    tables = [f"""{table_name}({",".join(tables[table_name])})""" for table_name in tables]
    tables = ", ".join(tables)
    prompt = get_prompt(tables, question)
    input_ids = tokenizer(prompt, max_length=1000, return_tensors="pt",truncation=True).input_ids
    return input_ids

def inference(question: str, tables: dict[str, List[str]]) -> str:
    input_data = prepare_input(question=question, tables=tables)
    input_data = input_data.to(model.device)
    outputs = model.generate(inputs=input_data,  max_length=1000)
    result = tokenizer.decode(token_ids=outputs[0], skip_special_tokens=True)
    return result
```

```python
from fuzzywuzzy import fuzz
from fuzzywuzzy import process


def get_closest_country(input_country, country_list):
    # Find the closest match to the input country from the country list
    best_match = process.extractOne(input_country, country_list)
    return best_match[0]  # Return the closest country name

def extract_data(sql):
    # From the genereated SQL retrieve the data and return both data and SQL query
    conn = sqlite3.connect("database_life_exp.db")
    cursor = conn.cursor() # open database

    # try the sql call
    try:
        if 'Country =' in sql: # Extract country from the SQL query (if applicable)
            cursor.execute("SELECT DISTINCT Country FROM Life_Expectancy")
            countries = [row[0] for row in cursor.fetchall()] # all countries in the data

            start = 0
            while 'Country = "' in sql[start:]:
                # Find the start and end of the current country in the SQL string
                country_start = sql.find('Country = "', start) + len('Country = "')
                country_end = sql.find('"', country_start)
                
                # Extract the misspelled country
                misspelled_country = sql[country_start:country_end]
                
                # Correct the country name
                correct_country = get_closest_country(misspelled_country, countries)
                
                # Replace the misspelled country with the corrected one
                sql = sql[:country_start] + correct_country + sql[country_end:]
                
                # Update the start position for the next occurrence
                start = country_end
            
        cursor.execute(sql) 
        rows = cursor.fetchall()
        
        # Check if rows are empty or contain only (None,)
        if rows == [(None,)]:
            rows = 'No data found for the query.'
        
    except Exception as e:
        print(f"Error: {e}") 
        rows  = 'The SQL query didnt go through'
    finally:
        conn.close()

    return sql,rows # return the sql and fetched data

    
```

```python
from collections import Counter
import re


def rouge_2(generated, reference,tokeniser):
    """Compute the ROUGE-2 score between a system output and a reference.

    Arguments:
        generated (str): The generated output, as a single string.
        reference (str): The reference to compare against, as a single string.

    Returns:
        The F1-score of the ROUGE-2 metric between system output and reference.
    """
    
    generated_doc = tokeniser(generated).input_ids
    reference_doc = tokeniser(reference).input_ids
    # Generate bigrams (pairs of consecutive words)
    generated_bigrams = [(generated_doc[i], generated_doc[i + 1]) for i in range(len(generated_doc) - 1)]
    reference_bigrams = [(reference_doc[i], reference_doc[i + 1]) for i in range(len(reference_doc) - 1)]

    # Count bigrams using Counter
    generated_bigram_set = Counter(generated_bigrams)
    reference_bigram_set = Counter(reference_bigrams)

    # Calculate overlap between system and reference bigrams
    overlap = generated_bigram_set & reference_bigram_set

    # Compute precision, recall, and F1-score
    if len(generated_bigram_set) > 0:
        precision = sum(overlap.values()) / sum(generated_bigram_set.values())
    else:
        precision = 0

    if len(reference_bigram_set) > 0:
        recall = sum(overlap.values()) / sum(reference_bigram_set.values())
    else:
        recall = 0

    # Compute F1-score
    if precision + recall > 0:
        f1_score = (2 * precision * recall) / (precision + recall)
    else:
        f1_score = 0

    return f1_score


```

```python


from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline

# Load the Llama-3.2-3B model and tokenizer
model_name = "meta-llama/Llama-3.2-3B"

# Load the model and tokenizer
llm_tokenizer = AutoTokenizer.from_pretrained(model_name)
llm_model = AutoModelForCausalLM.from_pretrained(model_name, device_map="auto", torch_dtype="auto")
```

```python

# Define the function for generating responses
def generate_response(user_input,table_desc):
    """
    Generate a response using a structured prompt format with system, analysis, user input.
    """
    # Construct the structured prompt
    prompt = f"""
                System: You are a Generator (Knowledge assistant) in a RAG system tasked with providing a concise and detailed analysis of the data retrieved based on the user's question in 2 sentences. Ensure clarity and focus on the key details. The retriever can make mistakes. 
                       
                Database description:{table_desc}
                
                User input: {user_input}

                Assistant:
                """
    
    # Tokenize the input
    inputs = llm_tokenizer(prompt, return_tensors="pt", truncation=True, max_length=1024).to(llm_model.device)

    # Generate the response
    outputs = llm_model.generate(
        **inputs,
        max_new_tokens=100,
        temperature=0.1, # low randomness in answers
        top_p=0.90,
        pad_token_id=llm_tokenizer.eos_token_id
    )

    # Decode the response
    response = llm_tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Extract the assistant's response (remove the prompt part)
    assistant_response =  response.split("Assistant:")[-1].strip()
    assistant_response = assistant_response.split("\n")[0]
    return assistant_response


def RAG(question,scheme,table_desc):
    """Produce an abtractive summary for a given text.
    Retrive data and generate an answer given a question
    
    """
    query = inference_tune(question, scheme)
    
    sql,data = extract_data(query)


    user_prompt =  f"""" User question: {question} 
                         Retrieved Data: {data}"""
    
    if data == []:
        return 'The SQL query didnt go through, failed to retrieve any data'
    else:
        response = generate_response(user_prompt,table_desc)
        return response 
```
